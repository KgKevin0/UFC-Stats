## Put everything togheter - For Loop

```python
first_name = []
last_name = []
nickname = []
height = []
weight = []
reach = []
stance = []
wins = []
loss = []
draw = []
link = []

# loop over every page from a-z
for x in range(97,123):
    url = f"http://www.ufcstats.com/statistics/fighters?char={chr(x)}&page=all"
    
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    results = soup.find("table", {"class": "b-statistics__table"}).find("tbody").find_all("tr")[1:]
    
    for i in results:
    
        # first name
        first_name.append(i.find_all("td")[0].get_text().strip())
    
        # last name
        last_name.append(i.find_all("td")[1].get_text().strip())
    
        # nickname
        nickname.append(i.find_all("td")[2].get_text().strip())
    
        # height
        height.append(i.find_all("td")[3].get_text().strip())
    
        # weight
        weight.append(i.find_all("td")[4].get_text().strip())
    
        # reach
        reach.append(i.find_all("td")[5].get_text().strip())
    
        # stance
        stance.append(i.find_all("td")[6].get_text().strip())
    
        # wins
        wins.append(i.find_all("td")[7].get_text().strip())
    
        # loss
        loss.append(i.find_all("td")[8].get_text().strip())
    
        # draw
        draw.append(i.find_all("td")[9].get_text().strip())
        
        # link
        link.append(i.find_all('a', href=True)[0]["href"])
```

# Pandas Dataframe - UFC Fighter list

```python
ufc_df = pd.DataFrame({"First Name": first_name, "Last Name": last_name, "Nickname": nickname, 
                       "Height (foot)": height, "Weight (lbs)": weight, "Reach (inch)": reach, "Stance": stance, 
                       "W": wins, "L": loss, "D": draw, "Link": link})
```

## Put everything togheter - For Loop

```python
SLpM = []
Str_Acc = []
SApM = []
Str_Def = []
TD_Avg = []
TD_Acc = []
TD_Def = []
Sub_Avg = []
Link2 = []

# loop over every individual fighter profile link in Link column
for link in ufc_df.Link:
    
    url2 = link
    response2 = requests.get(url2)
    soup = BeautifulSoup(response2.content, "html.parser")
    
    stats_result = []
    result1 = soup.find("div", {"class": "b-list__info-box-left"}).find_all("li")[0:4]
    result2 = soup.find("div", {"class": "b-list__info-box-left"}).find_all("li")[5:]
    stats_result = result1 + result2
    
    # Link
    Link2.append(url2)
    
    # SLpM - Significant Strikes Landed per Minute
    element = stats_result[0].find("i")
    if element:
        element.extract()
    
    SLpM.append(stats_result[0].get_text().strip())
    
    # Str. Acc - Significant Striking Accuracy
    element = stats_result[1].find("i")
    if element:
        element.extract()
    
    Str_Acc.append(stats_result[1].get_text().strip())
    
    # SApM - Significant Strikes Absorbed per Minute
    element = stats_result[2].find("i")
    if element:
        element.extract()
    
    SApM.append(stats_result[2].get_text().strip())
    
    # Str. Def - Significant Strike Defence (the % of opponents strikes that did not land)
    element = stats_result[3].find("i")
    if element:
        element.extract()
    
    Str_Def.append(stats_result[3].get_text().strip())
    
    # TD Avg - Average Takedowns Landed per 15 minutes
    element = stats_result[4].find("i")
    if element:
        element.extract()
    
    TD_Avg.append(stats_result[4].get_text().strip())
    
    # TD Acc - Takedown Accuracy
    element = stats_result[5].find("i")
    if element:
        element.extract()
    
    TD_Acc.append(stats_result[5].get_text().strip())
    
    # TD Def - Takedown Defense (the % of opponents TD attempts that did not land)
    element = stats_result[6].find("i")
    if element:
        element.extract()
    
    TD_Def.append(stats_result[6].get_text().strip())
    
    # Sub. Avg - Average Submissions Attempted per 15 minutes
    element = stats_result[7].find("i")
    if element:
        element.extract()
    
    Sub_Avg.append(stats_result[7].get_text().strip())
```

# Pandas DataFrame - Fighter performance stats

```python
# Create new Dataframe with more stats for every UFC fighter
ufc_df2 = pd.DataFrame({"Link": Link2, "SLpM": SLpM, "Str Acc (%)": Str_Acc, "SApM": SApM, "Str Def (%)": Str_Def, 
                        "TD Avg": TD_Avg, "TD Acc (%)": TD_Acc, "TD Def (%)": TD_Def, "Sub Avg": Sub_Avg})

ufc_df2
```
